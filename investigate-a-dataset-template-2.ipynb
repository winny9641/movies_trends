### Data Cleaning (Fixing data: 'genres')
> Every row for 'genres', data stored multiple genres by seperating with "|". Make new data frame with single genre.


```python
movies_df = pd.read_csv('movies_clean.csv')
```

```python
movies_df.head()
```

```python
single_df = movies_df[~movies_df['genres'].str.contains("\|")] #data that has single genre 
```

```python
single_df
```

```python
single_df.genres.str.contains("\|").value_counts(False) #check
```

```python
multi_df = movies_df[movies_df['genres'].str.contains("\|")] # data set with multiple genres
```

```python
movies_df.shape
```

```python
single_df.shape[0]+multi_df.shape[0] #check 
```

seperate the multiple genres into single genres form

```python
for index, row in multi_df.iterrows(): #itterate the row of the data with multiple genres data 
    genres = row.genres.split("|") # each row, collect the genres
    for i in genres:# each genres in the row
        new_row = row.replace(row.genres,i) # make the new row with one genre
        single_df = single_df.append(new_row, ignore_index= True) #appends to single genre data set
  
```

```python
single_df.duplicated().sum() #check
```

```python
single_df.to_csv('single_genres.csv',index_label = False) # save as csv file 
```

#### Year by Year Popular Genre

```python
popular_genres = pd.DataFrame(columns = ['year','popular_genre'])
```

```python
genre_list = []
```

```python
for year in years:
    year_df = single_genres[single_genres.release_year == year]
    genre_ = year_df.groupby('genres').mean().popularity.idxmax()
    genre_list.append(genre_)
    
    
    
    
```

```python
popular_genres.year = years
```

```python
popular_genres.popular_genre = genre_list
```

```python
popular_genres
```

```python
ax = popular_genres.popular_genre.value_counts().plot.bar();
ax.set_xlabel('Genres');
ax.set_ylabel('Countes');
ax.set_title('Yearly Highest Popular Genres');
```

Year_to_Year:Adventure, Animation,Fantasy,Crime are mostly ranked as year's most popular genre

### Research Question 4 (Which genres result high revenue year to year)


```python
rev_genres = pd.DataFrame(columns = ['year','high_rev_genre'])
```

```python
rev_ = []
```

```python
for year in years: # year to year
    year_df = single_genres[single_genres.release_year == year] #collect each year
    genre_ = year_df.groupby('genres').mean().revenue_adj.idxmax() #find the highest average genres in the year
    rev_.append(genre_) # append to data frame
```

```python
rev_genres.year = years
rev_genres.high_rev_genre = rev_
rev_genres
```

```python
ax = rev_genres.high_rev_genre.value_counts().plot.bar();
ax.set_xlabel('Genres');
ax.set_ylabel('Counts');
ax.set_title('Yearly highest revenue Genre ');
```

```python
adventure_por = rev_genres.high_rev_genre.value_counts().Adventure / rev_genres.high_rev_genre.value_counts().sum()
fantasy_por =  rev_genres.high_rev_genre.value_counts().Fantasy / rev_genres.high_rev_genre.value_counts().sum()
animation_por =  rev_genres.high_rev_genre.value_counts().Animation / rev_genres.high_rev_genre.value_counts().sum()
war_por =  rev_genres.high_rev_genre.value_counts().War / rev_genres.high_rev_genre.value_counts().sum()
family_por =  rev_genres.high_rev_genre.value_counts().Family / rev_genres.high_rev_genre.value_counts().sum()
others_por = 1- (adventure_por + fantasy_por + animation_por + war_por + family_por)
[adventure_por,fantasy_por,animation_por,war_por,family_por,others_por]
```
